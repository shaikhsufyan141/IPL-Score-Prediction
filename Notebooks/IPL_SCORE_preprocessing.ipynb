```python
data = pd.read_csv(r"C:\Users\shaik\Downloads\ipl.csv")
data.head()
```

```python
data.drop("mid",axis=1,inplace=True)
```

```python
data.drop(["batsman","bowler","striker","non-striker"],axis=1,inplace=True)
```

```python
data.head()
```

# REMOVING THE FIRST 5 OVER DATA IN EVERY MATCH

```python
data = data[data['overs']>=5.0]
```

# CONVERTING THE DATE COLUMN FROM STRING INTO DATETIME OBJECT

```python
from datetime import datetime
data['date'] = data['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

# CONVERTING THE CATEGORICAL DATA INTO NUMERICAL

```python
encoded_df = pd.get_dummies(data = data, columns=['bat_team','bowl_team'])
```
